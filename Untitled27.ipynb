```python
df_country=pd.read_excel('/content/Country-Code.xlsx',)
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
Country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
plt.pie(country_value[:3],labels=Country_names[:3],autopct='%1.3f%%')
```

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
rating
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] =(12,6)
sns.barplot(x="Aggregate rating",y="rating count",data=rating)
```

```python
sns.barplot(x="Aggregate rating",y="rating count",hue='Rating color',data=rating,palette=['white','red','orange','yellow','green','green'])
```
